```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt

urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placekitten.com/800/571"]


for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5 )
  image_2 =cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame =cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')

tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi X lebar = ", tinggi,"x",lebar)
cv2_imshow(image_2)

print("Garis")
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)

# #Tugas 2
#Garis vartikal
tinggi, lebar, _ = image.shape
image_rgb1 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_rgb2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for x in range(tinggi):
    image_rgb2[x, lebar // 2] = [255, 255, 255]


final_frame = cv.hconcat((image_rgb1, image_rgb2))
cv2_imshow(final_frame)


# Menggambar garis silang
image_rgb3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_rgb4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Mengimage garis X (silang)
cv.line(image_rgb4, (0, 0), (lebar, tinggi), (255, 255, 255), 2)  # Garis miring dari kiri atas ke kanan bawah
cv.line(image_rgb4, (0, tinggi), (lebar, 0), (255, 255, 255), 2)  # Garis miring dari kiri bawah ke kanan atas


final = cv.hconcat((image_rgb3, image_rgb4))
cv2_imshow(final)

# tugas 3
# Buat garis horisontal berwarna putih dibagian tengah gambar dengan panjang tertentu

# Ambil tinggi dan lebar gambar
tinggi, lebar, _ = image.shape


img1 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Panjang garis yang diinginkan (misalnya, setengah dari lebar gambar)
panjang_garis = lebar // 5

# Menggambar garis horizontal putih di tengah gambar
cv.line(img2, (0, tinggi // 4), (panjang_garis, tinggi // 4), (255, 255, 255), 4)  # Garis horizontal putih

# Menampilkan image hasilnya
final=cv.hconcat((img1, img2))
cv2_imshow(final)

# Tugas 3 Buat kotak menggunakan kumpulan pixel warna putih di sembarang tempat dalam gambar Ambil tinggi dan lebar gambar
tinggi, lebar, _ = image.shape

img3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
img4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Koordinat sudut kiri atas kotak (x, y)
x_sudut_kiri_atas = 100
y_sudut_kiri_atas = 100

# Ukuran kotak (panjang dan tinggi)
panjang_kotak = 200
tinggi_kotak = 150

# Mengimage kotak putih di image
cv.rectangle(img4, (x_sudut_kiri_atas, y_sudut_kiri_atas), (x_sudut_kiri_atas + panjang_kotak, y_sudut_kiri_atas + tinggi_kotak), (255, 255, 255), -1)

# Menampilkan image hasilnya

hasil = cv.hconcat((img3, img4))
cv2_imshow(hasil)


# Tugas kelompok
from PIL import Image
from google.colab import files

# Unggah file gambar
from google.colab import files
from IPython.display import Image

# Unggah file gambar
uploaded = files.upload()

# Nama file gambar yang diunggah
nama_file = list(uploaded.keys())[0]

nama_file= cv.imread(nama_file)
nama_file = cv.cvtColor(nama_file,cv.COLOR_BGR2RGB)
plt.imshow(nama_file)

nama_file[0:30,195:605] = [248,253,95]
nama_file[32:60,260:550] = [0,0,0]
plt.imshow(nama_file)

# Menampilkan gambar
Image(nama_file)
```
